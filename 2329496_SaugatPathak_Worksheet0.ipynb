```python
#TASK1
def length_conversion(value, conversion_type):
  #convert length between meter and feet

  if conversion_type == 'm_to_ft':
    return value * 3.28084 #1 meter = 3.28084 feet
  elif conversion_type == 'ft_to_m':
    return value / 3.28084 #1 foot = 0.3048 meter
  else:
    return None

def weigth_conversion(value, conversion_type):
 #convert length between kilogram and pounds

  if conversion_type == 'kg_to_lbs':
    return value * 2.20462 #1 kg = 2.20462 lbs
  elif conversion_type == 'lbs_to_kg':
      return value / 2.20462 #1 lbs = 0.453592 kg
  else:
    return None

def volume_conversion(value, conversion_type):

  #convert length between liter and gallon

  if conversion_type == 'L_to_gal':
    return value * 0.264172  # 1 liter = 0.264172 gallons
  elif conversion_type == 'gal_to_L':
      return value / 0.264172  # 1 gallon = 3.78541 liters
  else:
    return None


def main():

  #main function to run the program

  print("Welcome to the length conversion program! \n")
  print("Please select an option:")
  print("1. Convert Length")
  print("2. Convert Weight")
  print("3. Convert Volume \n")

  try:
    choice = int(input("Enter your Choice 1, 2 and 3: "))
    if choice not in [1,2,3]:
      print("Please enter number between 1, 2 and 3")
      return

    value = float(input("Enter value to convert: "))

    if choice == 1:
      print("1: Convert meter to feet")
      print("2. Convert feet to meter")
      length_choice = int(input("Enter Your choice (1/2): "))
      conversion_type = 'm_to_ft' if length_choice == 1 else 'ft_to_m'
      result = length_conversion(value, conversion_type)

    elif choice == 2:
      print("1: Convert meter to feet")
      print("2. Convert feet to meter")
      weight_choice = int(input("Enter Your choice (1/2): "))
      conversion_type = 'kg_to_lbs' if weight_choice == 1 else 'lbs_to_kg'
      result = weigth_conversion(value, conversion_type)

    elif choice == 3:
      print("1: Convert meter to feet")
      print("2. Convert feet to meter")
      volume_choice = int(input("Enter Your choice (1/2): "))
      conversion_type = 'L_to_gal' if volume_choice == 1 else 'gal_to_L'
      result = volume_conversion(value, conversion_type)

    print(f"converted value : {result:4f}")

  except ValueError:
    print("Invalid Input! please enter a numeric value")

if __name__ == '__main__':
  main()
```

# 4.2 Exercise on List Manipulation:

```python
#1 Extract Every other element

def extract_every_other(lst):

  #Extracts every other element from a list, starting from the first element.


  return lst[::2]

#2 Slice a sublist

def get_sublist(lst, start, end):

 # Extracts a sublist from a given list based on specified start and end indices.


  return lst[start:end+1]

#3 Reverse a List using slicing

def reverse_list(lst):

  #Reverses a list using slicing.


  return lst[::-1]

#4 Remove the first and last element

def remove_first_last(lst):

 # Removes the first and last elements of a list and returns the resulting sublist.

  return lst[1:-1]

#5. Get the first n element

def get_first_n(lst, n):

  #Extracts the first n elements from a list.

  return lst[:n]

#6. Extract Element from the end

def get_last_n(lst, n):

  #Extracts the last n elements from a list using slicing.

  return lst[-n:]

#7. Extract Element in reverse order

def reverse_skip(lst):

  #Extracts elements in reverse order, skipping one in between, starting from the second-to-last element.

  return lst[-2::-2]

#Example list
sample_list =  [1,2,3,4,5,6]

print("1. Extract every other element:", extract_every_other(sample_list))
print("2. Sublist (2 to 4):", get_sublist(sample_list, 2, 4))
print("3. Reversed list:", reverse_list(sample_list))
print("4. Without first & last:", remove_first_last(sample_list))
print("5. First 3 elements:", get_first_n(sample_list, 3))
print("6. Last 2 elements:", get_last_n(sample_list, 2))
print("7. Reverse skip:", reverse_skip(sample_list))
```

# 4.3 Exercise on Nested List:

```python
#1. flatten a Nested List

def flatten(lst):

  #Flattens a nested list into a single list.


  result = []
  for sublist in lst:
    if isinstance(sublist, list):
      result.extend(flatten(sublist))
    else:
      result.append(sublist)
  return result

#2. Accessing Nested List Elements
def access_nested_element(lst, indices):

  #Extracts a specific element from a nested list given its indices.


  element = lst
  for index in indices:
    element = element[index]
  return element

#3. Sum of All Elements in a Nested List

def sum_nested(lst):

  #Calculates the sum of all numbers in a nested list, regardless of depth.

  total = 0
  for item in lst:
    if isinstance(item, list):
      total += sum_nested(item)
    else:
      total +=item
  return total

# 4. Remove Specific Element from a Nested List

def remove_element(lst, elem):

  #Removes all occurrences of a specific element from a nested list.
  result = []
  for item in lst:
    if isinstance(item, list):
      filtered_list = remove_element(item, elem)
      if filtered_list:
        result.append(filtered_list)
    elif item != elem:
      result.append(item)
  return result

# 5. Find the Maximum Element in a Nested List

def find_max(lst):

  #Finds the maximum element in a nested list.

  max_value = float('-inf')
  for item in lst:
    if isinstance(item, list):
      max_value = max(max_value, find_max(item))
    else:
      max_value = max(max_value, item)
  return max_value

#6. Count Occurrences of an Element in a Nested List

def count_occurrences(lst, elem):

  #Counts the occurrences of a specific element in a nested list.
  count = 0
  for item in lst:
    if isinstance(item, list):
      count += count_occurrences(item, elem)
    elif item == elem:
      count +=1
  return count

#7. Flatten a List of Lists of Lists

def deep_flatten(lst):

  #Flattens a deeply nested list into a single list.

  result = []
  for item in lst:
    if isinstance(item, list):
      result.extend(deep_flatten(item))
    else:
      result.append(item)
  return result

# 8. Nested List Average

def average_nested(lst):

   #Calculates the average of all elements in a nested list.

  flat_list = deep_flatten(lst)
  return sum(flat_list) / len(flat_list) if flat_list else 0


#Example Input list
nested_list_1 = [[1,2],[3,4],[5]]
nested_list_2 = [[1,2,3],[4,5,6],[7,8,9]]
nested_list_3 = [[1,2],[3,[4,5]],6]
nested_list_4 = [[1,2],[3,2],[4,5]]
nested_list_5 = [[[1,2],[3,4]],[[5,6],[7,8]]]


print("1. Flatten a Nested List: ", flatten(nested_list_1))
print("2. Access Nested Element (index [1,2]): ", access_nested_element(nested_list_2, [1,2]))
print("3. Sum of All Elements:", sum_nested(nested_list_3))
print("4. Remove Element (elem=2):", remove_element(nested_list_4, 2))
print("5. Max Element in Nested List:", find_max(nested_list_3))
print("6. Count Occurrences (elem=2):", count_occurrences(nested_list_4, 2))
print("7. Deep Flatten:", deep_flatten(nested_list_5))
print("8. Average of Nested List:", average_nested([[1, 2], [3, 4], [5, 6]]))
```

`Problem 4: Matrix Operations`

```python
import numpy as np

A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])

# 1. Prove A * A⁻¹ = I
A_inv = np.linalg.inv(A)
identity_check = np.dot(A, A_inv)
print("A * A⁻¹ = Identity Matrix:\n", identity_check)

# 2. Prove AB ≠ BA
AB = np.dot(A, B)
BA = np.dot(B, A)
print("\nMatrix AB:\n", AB)
print("\nMatrix BA:\n", BA)
if not np.array_equal(AB, BA):
    print("\nAB is NOT equal to BA")

# 3. Prove (AB)ᵀ = Bᵀ * Aᵀ
AB_transpose = np.transpose(AB)
BT_AT = np.dot(B.T, A.T)
print("\n(AB)ᵀ:\n", AB_transpose)
print("\nBᵀ * Aᵀ:\n", BT_AT)
if np.array_equal(AB_transpose, BT_AT):
    print("\n(AB)ᵀ is equal to Bᵀ * Aᵀ")

# 4. Solve Linear Equations using Inverse Methods
# System of equations:
#  2x - 3y + z = -1
#  x - y + 2z = -3
#  3x + y - z = 9

A_matrix = np.array([[2, -3, 1],
                     [1, -1, 2],
                     [3, 1, -1]])

B_matrix = np.array([-1, -3, 9])

A_inv = np.linalg.inv(A_matrix)
solution = np.dot(A_inv, B_matrix)
print("\nSolution for Linear Equations (x, y, z):\n", solution)
```
